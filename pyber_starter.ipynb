```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
city_data_to_load = "c:/Users/cvargas/Desktop/Matplotlib_Challenge/05-Matplotlib/Instructions/Pyber/data/city_data.csv"
ride_data_to_load = "c:/Users/cvargas/Desktop/Matplotlib_Challenge/05-Matplotlib/Instructions/Pyber/data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
city_data.drop_duplicates('city', inplace=False)
merged_data = ride_data.merge(city_data, on='city', how='left')
merged_data['type'].unique()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
Urban = merged_data[merged_data.type == 'Urban']
Suburban = merged_data[merged_data.type == 'Suburban']
Rural = merged_data[merged_data.type == 'Rural']
# Build the scatter plots for each city types
UrbanAvg = Urban.groupby(['city']).mean()
UrbanTot = Urban.groupby(['city']).count()
UrbanDrv = Urban.groupby(['city']).sum()
SuburbanAvg = Suburban.groupby(['city']).mean()
SuburbanTot = Suburban.groupby(['city']).count()
SuburbanDrv = Suburban.groupby(['city']).sum()
RuralAvg = Rural.groupby(['city']).mean()
RuralTot = Rural.groupby(['city']).count()
RuralDrv = Rural.groupby(['city']).sum()

# Incorporate the other graph properties
UrbanBu = plt.scatter(UrbanTot['date'], UrbanAvg['fare'], s=UrbanDrv['driver_count'], color="LightCoral", alpha=0.7, edgecolors='Black', label='Urban')
SuburbanBu = plt.scatter(SuburbanTot['date'], SuburbanAvg['fare'], s=SuburbanDrv['driver_count'], alpha=0.7, color="LightSkyBlue", edgecolors='Black', label='Suburban')
RuralBu = plt.scatter(RuralTot['date'], RuralAvg['fare'], s=RuralDrv['driver_count'], color='Gold', alpha=0.7, edgecolors='Black', label='Rural')
# Create a legend
cuadro = plt.legend(handles=[UrbanBu, SuburbanBu, RuralBu], title="City Types", fontsize=8, loc="best")
cuadro.legendHandles[0]._sizes = [50]
cuadro.legendHandles[1]._sizes = [50]
cuadro.legendHandles[2]._sizes = [50]
# Incorporate a text label regarding circle size
plt.title('Pyber Ride Sharing Data')
plt.xlabel('Count of rides')
plt.ylabel('Avg ride price')
plt.text(43, 40, 'Note:\n Circle size correlates with driver count per city.')
plt.grid()
# Save Figure
plt.savefig('Matplotlib_Challenge.png')
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
Fare = merged_data.groupby(['type'])['fare'].sum()
label = ["Rural", "Suburban", "Urban"]
explode = (0, 0, 0.1)
colors = ["gold","lightskyblue","lightcoral"]
# Build Pie Chart
plt.pie(Fare, labels=label, autopct='%.1f%%', explode=explode, colors=colors, shadow=True, startangle=160)
plt.show()
# Save Figure
plt.savefig('Matplotlib_Challenge2.png')
```

## Total Rides by City Type

```python
# Calculate Ride Percents
Fare = merged_data.groupby(['type'])['fare'].count()
label = ["Rural", "Suburban", "Urban"]
explode = (0, 0, 0.1)
colors = ["gold","lightskyblue","lightcoral"]
# Build Pie Chart
plt.pie(Fare, labels=label, autopct='%.1f%%', explode=explode, colors=colors, shadow=True, startangle=160)
plt.show()
# Save Figure
plt.savefig('Matplotlib_Challenge3.png')
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
Fare = city_data.groupby(['type'])['driver_count'].sum()
label = ["Rural", "Suburban", "Urban"]
explode = (0, 0, 0.1)
colors = ["gold","lightskyblue","lightcoral"]
# Build Pie Chart
plt.pie(Fare, labels=label, autopct='%.1f%%', explode=explode, colors=colors, shadow=True, startangle=160)
plt.show()
# Save Figure
plt.savefig('Matplotlib_Challenge4.png')
```
